### Criando um gráfico de Dispersão com o Matplotlib


```python
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
ano = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
quantidade = [112, 145, 175, 210, 237, 254, 266, 278, 295, 305, 312] 
```

```python
cores = plt.get_cmap('Accent').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.figure(figsize=(10, 5))

plt.scatter(ano, quantidade, marker='o', s=quantidade, c=quantidade, cmap='Oranges')

plt.xticks(ano, rotation=55, fontsize=10, fontweight='bold')

plt.xlabel('Ano', font='Verdana', fontweight='bold')
plt.ylabel('Quantidade Vendida (em milhar)', font='Verdana', fontweight='bold')
plt.title('Vendas de Notebooks nos últimos 11 anos', font='Verdana', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='-.')

plt.colorbar()

plt.show()
```
