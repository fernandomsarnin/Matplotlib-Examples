```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [185, 190, 178, 192, 188, 196, 194, 198, 200, 202, 205, 208]

marcadores = np.where(vendas == np.min(vendas), 'p', '')

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores):
    if marcador == 'p':
        cor = 'red'
        tamanho = 80
    else:
        cor='forestgreen'
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-5, 5), textcoords='offset points')
    

plt.axvspan(6,11, alpha=0.2, color=cores[1])

plt.text(6, 178, 'Recuperação de Crescimento', color='green', fontsize=9, fontweight='bold')

plt.box(False)
plt.tick_params(axis='y', length=0)
plt.tick_params(axis='x', length=0)

plt.yticks([])

#plt.xlabel('Mês')
plt.ylabel('Vendas (Milhares)')
plt.title('Crescimento das Vendas Durante o Ano de 2021')


plt.show
```
