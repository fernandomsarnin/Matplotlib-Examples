### Crianção de gráfico de Pizza com dados fictícios utilizando a biblioteca Matplotlib.

```python
empresa = ['Columbian', "Graphc Style", "New Gool",	"Generation Blue",	"Enterpraise One"]
receita = [1680, 1320, 858, 350, 452]
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

destaques = [0.1, 0.05, 0, 0,  0]

plt.pie(receita, 
        labels=empresa, 
        autopct='%.1f', 
        explode=destaques, 
        shadow=True)
plt.title('Faturamento Anual das Empresas do Grupo', 
          font='Calibri', 
          fontsize=16,
          fontweight='bold')

plt.show()
```
