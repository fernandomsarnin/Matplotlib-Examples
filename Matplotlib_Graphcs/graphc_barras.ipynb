### Construção de um gráfico de Barras (Bar) com o Matplotlib

```python
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
ano = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
faturamento = [35.98, 48.13, 58.78, 62.85, 67.69, 58.57, 78.00, 81.48, 85.5]

cores = plt.get_cmap('Paired').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.figure(figsize=(10, 5))

barras = plt.bar(ano, faturamento)

plt.bar_label(barras, rotation=0, font='Calibri', fontsize=12)

plt.xticks(ano, rotation=55, fontsize=11, fontweight='bold')

plt.title('Faturamento da Comtex Corporation nos últimos 08 anos',
          font='Calibri',
          fontsize=18,
          fontweight='bold')


plt.xlabel('Ano', font='Calibri', fontsize=11.5, fontweight='bold')
plt.ylabel('Receita (Milhões de Reais)', font='Calibri', fontsize=12, fontweight='bold')

plt.show()
```
